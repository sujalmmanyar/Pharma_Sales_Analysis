```python
month_map = {'January 24':'Jan 2024','Jan-2024':'Jan 2024','JAN-2024':'Jan 2024'}
```

```python
df['Month'] = df['Month'].replace(month_map)
```

```python
df['Month'] = pd.to_datetime(df['Month'])
```

```python
df['Month'] = df['Month'].dt.strftime('%b-%Y')
```

```python
df['Stockist Name'] = df['Stockist Name'].replace('A-Z PHARMA','A-Z Pharma')
```

```python
df['Product Name'] = df['Product Name'].str.strip().str.lower()
```
